## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the data by Item ID and Item Name to calculate the count of purchases
item_group = purchase_data2.groupby(['Item ID', 'Item Name'])
purchase_count_by_item = item_group.size().reset_index(name='Purchase Count')

# Sort the items in descending order based on the count of purchases
sorted_items = purchase_count_by_item.sort_values('Purchase Count', ascending=False)

most_popular_item = sorted_items.iloc[0]

# Filter the original DataFrame to include only the rows for the most popular item
popular_item_data = purchase_data2[(purchase_data2['Item ID'] == most_popular_item['Item ID']) | (purchase_data2['Item Name'] == most_popular_item['Item Name'])]
popular_item_data
```

```python
# Group the data by Item ID and Item Name and perform computation
item_group = purchase_data2.groupby(['Item ID', 'Item Name'])

# Perform calculations for purchase count, item price, and total purchase value
purchase_count = item_group['Purchase ID'].count()
item_price = item_group['Price'].mean()
total_purchase_value = item_group['Price'].sum()

# Create a new DataFrame to store the calculations
item_summary = pd.DataFrame({
    'Purchase Count': purchase_count,
    'Item Price': item_price,
    'Total Purchase Value': total_purchase_value
})

# Print the item summary
item_summary
```
